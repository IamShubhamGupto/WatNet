```python
import os
os.chdir('..')
import glob
import numpy as np
import tensorflow as tf
from utils.tiff_io import readTiff
from dataloader.path_io import crop_patch
from dataloader.tfrecord_io import image_example
```

```python
path_tra_tfrecord = 'data/tfrecord-s2/tra_data.tfrecords'
path_val_tfrecord_patch = 'data/tfrecord-s2/val_data_patch.tfrecords'
path_val_tfrecord_scene = 'data/tfrecord-s2/val_data_scene.tfrecords'
tra_scene_paths = sorted(glob.glob('data/dataset-s2/tra_scene/*.tif'))
tra_truth_paths = sorted(glob.glob('data/dataset-s2/tra_truth/*.tif'))
tra_pair_data = list(zip(tra_scene_paths, tra_truth_paths))
print('tra_data length:', len(tra_pair_data))
val_scene_paths = sorted(glob.glob('data/dataset-s2/val_scene/*.tif'))
val_truth_paths = sorted(glob.glob('data/dataset-s2/val_truth/*.tif'))
val_pair_data = list(zip(val_scene_paths, val_truth_paths))
print('val_data length:', len(val_pair_data))

```

```python
# Val data: Write to a `_scene.tfrecords` file.
with tf.io.TFRecordWriter(path_val_tfrecord_scene) as writer:
    for path_scene, path_truth in val_pair_data:
        print(path_scene)
        _,scene = readTiff(path_scene)
        _,truth = readTiff(path_truth)
        scene = np.clip(scene/10000,0,1)  
        tf_example = image_example(scene, truth)
        writer.write(tf_example.SerializeToString())
```
