```python
import os
os.chdir('..')
import tensorflow as tf
import numpy as np
from utils.tiff_io import readTiff
from dataloader.tfrecord_io import image_example
import glob
```

```python
path_tra_tfrecord = 'data/tfrecord-s2/tra_data_tmp.tfrecords'
scene_paths = sorted(glob.glob('data/dataset-s2/tra_scene/*.tif'))
truth_paths = sorted(glob.glob('data/dataset-s2/tra_truth/*.tif'))
tra_pair_data = list(zip(scene_paths,truth_paths))
```

```python
# Write to a `.tfrecords` file.
with tf.io.TFRecordWriter(path_tra_tfrecord) as writer:
    for path_scene, path_truth in tra_pair_data:
        print(path_scene)
        _,scene = readTiff(path_scene)
        _,truth = readTiff(path_truth)
        scene = np.clip(scene/10000,0,1)  
        tf_example = image_example(scene, truth)
        writer.write(tf_example.SerializeToString())
```
